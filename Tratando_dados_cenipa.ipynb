```
import pandas as pd
import glob as glob
import matplotlib.pyplot as plt
```

```
arquivos= glob.glob("/content/drive/My Drive/Colab Notebooks/BasesCenipa/ficheiro*.csv")
arquivos
```

```
base={}
n=1
colunas = list(range(1,42,1))
for arquivo in arquivos:
  ler_arquivo=pd.read_csv(arquivo, usecols=colunas)
  print(f'Arquivo {arquivo} lido.')
  n=n+1
  base[str(n)]=ler_arquivo
```

```
cenipa_df=pd.concat(base,ignore_index=True)
```

```
cenipa_df.to_csv('/content/drive/My Drive/Colab Notebooks/BasesCenipa/cenipa_df.csv')
```

```
cenipa_df.columns=['tipo_reporte','matricula','data_evento','operador','fabricante','modelo','tipo_aviacao','motor_fabricante','motor_modelo','hora_local','hora_gmt','parte_dia','icao','pista','lado','asa','coordenada','altura_ft','velocidade_kt','fase_voo','parte_atingida_danificada','outra_parte_atingida1','numero_motor','numero_helice','ingestao','ingestao_multipla','efeito_voo','outro_efeito_voo','condicao_ceu','precipitacao','especie_tamanho','quantidade_avistamento_atingida','amostra_coletada_enviadaid_fotografada_enviadaID','danos_prejuizo','aeronave_indisponivel_horas','custos_direto','custos_indireto','tripulacao_alertada','funcao','classificacao_ocorrencia','data_reporte']
cenipa_df['data_evento']=pd.to_datetime(cenipa_df['data_evento'])
cenipa_df.index=pd.to_datetime(cenipa_df['data_evento'])
```

```
cenipa_por_ano=cenipa_df.groupby(by=[cenipa_df.index.year, cenipa_df.index.month,'icao','tipo_reporte'])[('funcao','fase_voo')].agg('count')
cenipa_por_ano.sort_values(by='funcao',ascending=False).head(50)
```

```
cenipa_por_ano_df=pd.DataFrame(cenipa_por_ano)
cenipa_por_ano_df
```
